## Visualization & Explanatory Analysis

### Pertanyaan 1: What is the overall distribution of PM2.5 and PM10 levels across all stations?

```python
# Plot the distribution of PM2.5 and PM10
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
sns.histplot(combined_df['PM2.5'].dropna(), kde=True, bins=30, color='blue')
plt.title('Distribution of PM2.5')

plt.subplot(1, 2, 2)
sns.histplot(combined_df['PM10'].dropna(), kde=True, bins=30, color='orange')
plt.title('Distribution of PM10')

plt.tight_layout()
plt.show()
```

### Pertanyaan 2: How do meteorological factors (temperature, pressure, dew point, wind speed) correlate with pollutant levels (PM2.5, PM10, NO2, SO2, CO, O3)?

```python
# List of meteorological factors and pollutants
met_factors = ['TEMP', 'PRES', 'DEWP', 'WSPM']
pollutants = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

# Correlation matrix
corr_matrix = combined_df[met_factors + pollutants].corr()

# Heatmap of correlations
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation between Meteorological Factors and Pollutants')
plt.show()
```

### Pertanyaan 3: Which city/station has the highest levels of pollution over the dataset period?

```python
# Calculate mean pollutant levels per station
station_means = combined_df.groupby('station')[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()

# Sort by PM2.5 levels
top_polluted_stations = station_means.sort_values(by='PM2.5', ascending=False).head(10)

# Plot top 10 polluted stations
plt.figure(figsize=(10, 6))
sns.barplot(x=top_polluted_stations.index, y=top_polluted_stations['PM2.5'], palette='Reds')
plt.xticks(rotation=45)
plt.title('Top 10 Stations with Highest PM2.5 Levels')
plt.ylabel('PM2.5 Level')
plt.show()
```

### Pertanyaan 4: How does air quality change during different seasons (e.g., summer, winter)?

```python
# Define a function to assign seasons based on months
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Add season column
combined_df['season'] = combined_df['month'].apply(get_season)

# Group by season and calculate mean pollutant levels
season_means = combined_df.groupby('season')[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()

# Plot seasonal pollution levels
plt.figure(figsize=(10, 6))
season_means.plot(kind='bar', figsize=(10,6), stacked=True, cmap='Spectral')
plt.title('Average Pollutant Levels by Season')
plt.ylabel('Pollutant Level')
plt.show()
```

### Pertanyaan 5: What are the time trends (daily, monthly, yearly) for different pollutants across the stations?

```python
# Create a datetime column
combined_df['date'] = pd.to_datetime(combined_df[['year', 'month', 'day', 'hour']])

# Group by date and calculate daily averages for pollutants
daily_trends = combined_df.groupby(combined_df['date'])[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']].mean()

# Plot daily trends
plt.figure(figsize=(12, 8))
daily_trends.plot(figsize=(12,8))
plt.title('Daily Trends of Pollutants Over Time')
plt.ylabel('Pollutant Level')
plt.xlabel('Date')
plt.show()
```

## Analisis Lanjutan (Opsional)

**RFM Analysis**

Karena dataset kita terkait dengan kualitas udara dan bukan perilaku pembelian, RFM Analysis bisa diadaptasi untuk mengelompokkan stasiun berdasarkan Recency, Frequency, dan Monetary. Dalam konteks ini:

Recency: Dihitung sebagai jumlah hari sejak terakhir kali data diambil dari stasiun tertentu.
Frequency: Jumlah observasi data yang dikumpulkan oleh stasiun.
Monetary: Karena tidak ada pengeluaran langsung, kita bisa mengganti dengan pengukuran rata-rata polusi (PM2.5 atau PM10) yang diamati oleh stasiun.

```python
import pandas as pd

# Load dataset
combined_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/Air-quality-dataset/combined_data_airquality.csv')

# Menangani tanggal dan mengelompokkan berdasarkan stasiun
combined_df['datetime'] = pd.to_datetime(combined_df[['year', 'month', 'day', 'hour']])
current_date = combined_df['datetime'].max()

# Recency: Waktu sejak pengamatan terakhir
recency_df = combined_df.groupby('station')['datetime'].max().reset_index()
recency_df['Recency'] = (current_date - recency_df['datetime']).dt.days

# Frequency: Jumlah pengamatan oleh stasiun
frequency_df = combined_df.groupby('station')['datetime'].count().reset_index()
frequency_df.columns = ['station', 'Frequency']

# Monetary: Rata-rata level PM2.5 yang diamati stasiun
monetary_df = combined_df.groupby('station')['PM2.5'].mean().reset_index()
monetary_df.columns = ['station', 'Monetary']

# Gabungkan data Recency, Frequency, dan Monetary
rfm_df = recency_df.merge(frequency_df, on='station').merge(monetary_df, on='station')

# Tampilkan hasil RFM Analysis
rfm_df
```
